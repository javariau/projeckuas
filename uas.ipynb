```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Tesla_stock_data.csv')
df
```

```python
stast = df.describe()
print(stast)
```

```python
import seaborn as sns 
sns.barplot(data = df , x  = 'Date' , y = 'Close')
plt.xticks(rotation  = 45)
plt.tight_layout()
plt.show()
```

```python
df.plot()
plt.tight_layout()
```

```python
fig, ax = plt.subplots(nrows= 3,ncols= 2,figsize = (15,11))

ax[0,0].plot(df['Date'],df['Close'],marker = "o")
ax[0,0].set_title('')
ax[0,0].tick_params(axis = 'x',rotation = 45)

ax[1,0].plot(df['Date'],df['Open'],marker = "o")
ax[1,0].set_title('')
ax[1,0].tick_params(axis = 'x',rotation = 45)

ax[2,0].plot(df['Date'],df['Low'],marker = "o")
ax[2,0].set_title('')
ax[2,0].tick_params(axis = 'x',rotation = 45)

ax[0,1].plot(df['Date'],df['High'],marker = "o")
ax[0,1].set_title('')
ax[0,1].tick_params(axis = 'x',rotation = 45)

ax[1,1].plot(df['Date'],df['Volume'],marker = "o")
ax[1,1].set_title('')
ax[1,1].tick_params(axis = 'x',rotation = 45)

ax[2,1].axis('off')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
