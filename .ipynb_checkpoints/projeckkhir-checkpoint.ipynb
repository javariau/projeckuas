```python
df = pd.read_csv('/kaggle/input/global-economic-indicators-20102025/world_bank_data_2025.csv')
```

```python
def load_and_inspect_data(df):
    """
    Fonction pour inspecter les données de base
    """
    print("📋 INFORMATIONS GÉNÉRALES SUR LE DATASET")
    print("-" * 50)
    print(f"📊 Dimensions: {df.shape[0]} lignes × {df.shape[1]} colonnes")
    print(f"📅 Période couverte: {df['year'].min()} - {df['year'].max()}")
    print(f"🌎 Nombre de pays: {df['country_name'].nunique()}")
    print(f"📈 Variables économiques: {df.shape[1] - 3}")
    
    return df
```

```python
load_and_inspect_data(df)
```

```python
def data_quality_assessment(df):
    """
    Évaluation de la qualité des données
    """
    print("ÉVALUATION DE LA QUALITÉ DES DONNÉES")
    print("-" * 50)
    
    # Calcul des valeurs manquantes
    missing_data = pd.DataFrame({
        'Colonne': df.columns,
        'Valeurs_manquantes': df.isnull().sum(),
        'Pourcentage_manquant': (df.isnull().sum() / len(df)) * 100
    }).sort_values('Pourcentage_manquant', ascending=False)
    
    print(missing_data)
    
    # Visualisation des valeurs manquantes
    plt.figure(figsize=(15, 8))
    sns.heatmap(df.isnull(), cbar=True, yticklabels=False, cmap='viridis')
    plt.title('Carte des Valeurs Manquantes par Variable', fontsize=16, fontweight='bold')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
    
    return missing_data
```
