```python
import pandas as pd
```

```python
df = pd.read_csv('Tesla_stock_data.csv')
df
```

```python
stats = df.describe()
print(stats)
```

```python
import seaborn as sns
sns.barplot( data = df, x = 'Date', y = 'Close')
plt.xticks(rotation = 45)
plt.tight_layout()
```

```python
df["Date"].value_counts().plot.pie()
plt.legend()
plt.tight_layout()
```

```python
df.plot()
```

```python
fig,ax =plt.subplots(nrows = 3 , ncols =2 , figsize=(14,10))

ax[0,0].plot(df['Date'],df['Close'],marker = 'o')
ax[0,0].set_title("tutup")
ax[0,0].tick_params(axis='x', rotation=45)

ax[1,0].plot(df['Date'],df['High'],marker = 'o' )
ax[1,0].set_title("harga tinggi")
ax[1,0].tick_params(axis='x', rotation=45)

ax[2,0].plot(df['Date'],df['Low'],marker = 'o')
ax[2,0].set_title("harga turun")
ax[2,0].tick_params(axis='x', rotation=45)

ax[0,1].plot(df['Date'],df['Open'],marker = 'o')
ax[0,1].set_title("buka")
ax[0,1].tick_params(axis='x', rotation=45)

ax[1,1].plot(df['Date'],df['Volume'],marker = 'o')
ax[1,1].set_title("volume")
ax[1,1].tick_params(axis='x', rotation=45)

ax[2,1].axis('off')

plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
